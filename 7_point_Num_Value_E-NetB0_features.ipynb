```python
import pandas as pd
import numpy as np
import glob
import shutil
import os
from sklearn import linear_model
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
val_src_path = "dataset/7-points/Validation_outputs_from_7-networks_features/EfficientNetB0/"
meta_path =  "dataset/7-points/"
```

```python
meta_val = pd.read_csv(meta_path+"meta.csv", sep=",")
meta_val = meta_val[['case_num','seven_point_score']]

val_index_cv = pd.read_csv(meta_path+"valid_indexes.csv", sep=",")
val_index_ = val_index_cv.values

indices_val = []

for i in range(val_index_.shape[0]):
	indices_val.append(val_index_[i,0])
		
meta_val = meta_val.loc[indices_val]

val_data = meta_val.values

y_train = val_data[:,1]
print(y_train.shape)
```

```python
meta_test = pd.read_csv(meta_path+"meta.csv", sep=",")
meta_test = meta_test[['case_num','seven_point_score']]

test_index_cv = pd.read_csv(meta_path+"test_indexes.csv", sep=",")
test_index_ = test_index_cv.values

indices_test = []

for i in range(test_index_.shape[0]):
	indices_test.append(test_index_[i,0])
	
meta_test = meta_test.loc[indices_test]
test_data = meta_test.values

y_test = test_data[:,1]
print(y_test.shape)
```

```python
df_c1_val = pd.read_csv(val_src_path+"point-1_effB0_val_features.csv", sep=",", header=None)
df_c2_val = pd.read_csv(val_src_path+"point-2_effB0_val_features.csv", sep=",", header=None)
df_c3_val = pd.read_csv(val_src_path+"point-3_effB0_val_features.csv", sep=",", header=None)
df_c4_val = pd.read_csv(val_src_path+"point-4_effB0_val_features.csv", sep=",", header=None)
df_c5_val = pd.read_csv(val_src_path+"point-5_effB0_val_features.csv", sep=",", header=None)
df_c6_val = pd.read_csv(val_src_path+"point-6_effB0_val_features.csv", sep=",", header=None)
df_c7_val = pd.read_csv(val_src_path+"point-7_effB0_val_features.csv", sep=",", header=None)


data_val= pd.concat([df_c1_val,df_c2_val,df_c3_val,df_c4_val,df_c5_val,df_c6_val,df_c7_val],axis = 1)
print(data_val)

X_train = data_val.values
print(X_train.shape)
```

```python
test_src_path = "dataset/7-points/Testing_outputs_from_7-network_features/EfficientNetB0/"
```

```python
df_c1_test = pd.read_csv(test_src_path+"point-1_effB0_test_features.csv", sep=",", header=None)
df_c2_test = pd.read_csv(test_src_path+"point-2_effB0_test_features.csv", sep=",", header=None)
df_c3_test = pd.read_csv(test_src_path+"point-3_effB0_test_features.csv", sep=",", header=None)
df_c4_test = pd.read_csv(test_src_path+"point-4_effB0_test_features.csv", sep=",", header=None)
df_c5_test = pd.read_csv(test_src_path+"point-5_effB0_test_features.csv", sep=",", header=None)
df_c6_test = pd.read_csv(test_src_path+"point-6_effB0_test_features.csv", sep=",", header=None)
df_c7_test = pd.read_csv(test_src_path+"point-7_effB0_test_features.csv", sep=",", header=None)

data_test= pd.concat([df_c1_test,df_c2_test,df_c3_test,df_c4_test,df_c5_test,df_c6_test,df_c7_test],axis = 1)
print(data_test)

X_test = data_test.values

print(X_test.shape)
```

```python
clf = RandomForestRegressor()
clf.fit(X_train[:,:2],y_train)
y_pred = clf.predict(X_test[:,:2])

```

```python
# The mean squared error
print('Mean squared error: %.2f'
      % mean_squared_error(y_test, y_pred))
```

```python
lr = linear_model.LinearRegression(normalize=True)
lr.fit(X_train,y_train)
y_pred = lr.predict(X_test)
```

```python
print('Coefficients: \n', lr.coef_)
```

```python
# The mean squared error
print('Mean squared error: %.2f'
      % mean_squared_error(y_test, y_pred))
```

```python
# The coefficient of determination: 1 is perfect prediction
print('Coefficient of determination: %.2f'
      % r2_score(y_test, y_pred))
```

```python
print(lr.score(X_test, y_test))
```
